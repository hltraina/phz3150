```python
#Problem 1
print("Problem 1:")
#Creation of arrays
x = np.arange(1, 100.5, 0.5)
y = np.log10(x)
z = np.sqrt(x)

plt.figure( figsize = (10,10) )
plt.plot  ( x, y, color = 'purple' , linewidth = 4, linestyle = 'dashdot' , label = "y(x)") 
plt.plot  ( x, z , color = 'teal', linewidth = 4, linestyle = 'dashdot', label = "z(x)" ) 
plt.xlabel( 'x values'                 , fontsize = 15 )
plt.ylabel( 'x and y values' , fontsize = 15 )
plt.title( 'Y(x) and Z(x) comparison' , fontsize = 20 )
plt.legend(fontsize = 10)
plt.xticks(np.arange(0,101, 10))
plt.yticks(np.arange(-1,2,0.5))
           
```

```python
print("Problem 3: ")
def word_to_number(word):
    """This function takes as input an English word that corresponds to a number from 0 to 20"""
    word_dict = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4, 'five': 5, 'six':6, 'seven':7, 'eight': 8, 'nine':9,
        'ten':10, 'eleven':11, 'twelve':12, 'thirteen':13, 'fourteen':14, 'fifteen':15, 'sixteen':16, 'seventeen':17,
        'eighteen':18, 'nineteen':19, 'twenty':20 }
    return word_dict[word]

print('Three is:', word_to_number('three'))
print('Seven is:', word_to_number('seven'))
print('Twelve is:', word_to_number('twelve'))
print('Sixteen is:', word_to_number('sixteen'))
print('Nineteen is:', word_to_number('nineteen'))
```
