```python
#Problem 1
print("Problem 1: ")
sentence = 'Deleting an item from a list or array while iterating over it is a Python problem that is well known to any experienced software developer'
max_word = ''
split = sentence.split() #created a variable to make references the sentence.split easier in the for/while loops
#printing each word: commented out after confirming the sentence displays as intended.
#for word in sentence.split() :
    #print(word)

##for loop:
for word in sentence.split():
    if len(word) > len(max_word):
        max_word = word
print(max_word)
```

```python
##while loop:
i=0 
max_word= "" 
while i<len(split):
    if len(split[i]) > len(max_word):
        max_word = split[i]
    i=i+1
    
print(max_word)
```

```python
print("Problem 4: ")
my_data = np.arange( 1, 26 ).reshape( (5, 5) )
print(my_data, "\n")
numbers = my_data[:,3]
print(numbers)

maximum = my_data[0, 0]
for row in my_data:
    for value in row: #uses value as an index in this secondary for loop
        if value > maximum:
            maximum = value
print(maximum)

total = 0
count = 0
for row in my_data:
    for value in row:
        total = value + total #is a sum of all the values in the array
        count = count + 1 #counts the number of elements in the array
        mean = total / count #uses the  function to calculate the mean
print(mean)

#a MUCH faster method of detecting max and mean is using numpy.
print(np.max(my_data), np.mean(my_data))
```
