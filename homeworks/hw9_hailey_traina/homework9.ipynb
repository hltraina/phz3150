```python
print("Question 3: ")
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from hw9_support_funct import piston

#part a
Values = np.linspace(1, .1, 40)
initial_pressure = 1.0
initial_volume = 1.0
initial_temperature = 300.0
gamma = 1.4

result = piston(Values, initial_pressure, initial_volume, initial_temperature, gamma)
print(result)

#part b
figure = plt.figure(figsize = (10,10))
ax = figure.add_subplot(111, projection = '3d')
ax.plot(result[:, 0], result[:, 1], result[:, 2], label = 'Adiabatic Expansion', linewidth = 2)
ax.view_init(elev=10, azim=35)
ax.set_xlabel('Pressure')
ax.set_ylabel('Volume')
ax.set_zlabel('Temperature')
ax.set_title('Adiabatic Expansion of Piston')
ax.legend()
plt.savefig('Piston_expansion.png')
plt.show()
```
