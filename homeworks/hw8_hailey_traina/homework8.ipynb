```python
#Question 2
class graduation_presents:
    def __init__(self, a, b, c, gift_numbers, wrapping_paper): #initiation function!
            self.a = a
            self.b = b
            self.c = c
            self.gift_numbers = gift_numbers
            self.wrapping_paper = wrapping_paper
    
    def surface_area(self):
        #Calculates the surface area of the box.
        area = 2 * (self.a * self.b + self.b * self.c + self.c * self.a)
        return area
    
    def volume(self):
        #Calculates the volume of the box
        vol = self.a * self.b * self.c
        return vol
    
    def enough_paper(self):
        #Check if there is enough wrapping paper for the box
        paper = self.surface_area() <= self.wrapping_paper
        return paper
    
    def fits(self):
        #Check if the gift can fit in the box
        gift_vol = 25
        vol_needed = self.gift_numbers * gift_vol
        return vol_needed <= self.volume()
    
    def print_gift(self): #the results that combines all information to determine if there is enough paper and room!
        if self.enough_paper and self.fits():
            print("You have enough wrapping papper to wrap your entire box!")
            print("All of your gifts fit in the box!")
        elif self.enough_paper():
            print("You have enough wrapping paper to wrap your box! Unfortunately, your gifts don't fit...")
        elif self.fits():
            print("Your gifts fit in the box! Unfortunately, you do not have a wrapping paper for your box...")
        else:
            print("You do not have enough wrapping paper and your gifts do not fit in the box :/")

test_case1 = graduation_presents(10, 10, 10, 15, 800)
test_case2 = graduation_presents(10, 10, 10, 55, 1000)
test_case3 = graduation_presents(100, 20, 10, 55, 300)

print("Test Case 1:")
test_case1.print_gift()

print("\nTest Case 2:")
test_case2.print_gift()

print("\nTest Case 3:")
test_case3.print_gift()            
            
            
```
